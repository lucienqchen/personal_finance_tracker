```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import math
plt.style.use("ggplot")
```

```python
stock = "TGT"
df = pdr.get_data_yahoo(stock)
df.tail()
```

```python
historical_prices = df.Close
historical_prices[:5]
```

```python
moving_average = historical_prices.rolling(window=20).mean()
for i in range(len(moving_average)):
    if (math.isnan(moving_average[i])):
        moving_average[i] = historical_prices[i]
moving_average
```

```python
historical_prices = pd.DataFrame(historical_prices)
historical_prices["Moving Average"] = moving_average
plt.figure(figsize=(10, 6))
plt.title("Moving Average of {}".format(stock))
plt.plot(historical_prices)
```
